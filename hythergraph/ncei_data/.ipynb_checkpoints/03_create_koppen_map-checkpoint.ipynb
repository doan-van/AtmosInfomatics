```python
def koppen_beck(precip, temp, south):
    # pre-calculations
    MAT = temp.sum() / 12  # mean air temperature
    
    MAP = precip.sum()  #  sum precipitation
    Pdry = precip.min()
    Tcold = temp.min()
    Thot = temp.max()
    
    Tmon10 = 0
    for ta in temp:
        if ta > 10: Tmon10 += 1

    if bool(south):   # southern hemisphere, winter from the 3rd to 9th month
        #print(south)
        he = "S"
        if precip[3:9].sum() > 0.7 * MAP:
            Pth = 2 * MAT
        elif np.concatenate((precip[0:3], precip[9:12])).sum() > 0.7 * MAP:  # summer
            Pth = 2 * MAT + 28
        else:
            Pth = 2 * MAT + 14
        Pwdry = precip[3:9].min()
        Pwwet = precip[3:9].max()
        Psdry = np.concatenate((precip[0:3], precip[9:12])).min()
        Pswet = np.concatenate((precip[0:3], precip[9:12])).max()
        
    else:  # northern hemisphere, summer from the 3rd to 9th month
        he = "N"
        if np.concatenate((precip[0:3], precip[9:12])).sum() > 0.7 * MAP:
            Pth = 2 * MAT
        elif precip[3:9].sum() > 0.7 * MAP:  # summer
            Pth = 2 * MAT + 28
        else:
            Pth = 2 * MAT + 14
        Psdry = precip[3:9].min()
        Pswet = precip[3:9].max()
        Pwdry = np.concatenate((precip[0:3], precip[9:12])).min()
        Pwwet = np.concatenate((precip[0:3], precip[9:12])).max()


    # classification conditionals
    if MAP < 10 * Pth:
        koppenClass = "B"
        if MAP < 5 * Pth:
            koppenClass = koppenClass + "W"
        else:
            koppenClass = koppenClass + "S"
        if MAT >= 18:
            koppenClass = koppenClass + "h"
        else:
            koppenClass = koppenClass + "k"
   
    # Tropical climates have an average temperature of 18 deg C or higher every month of the year, with significant precipitation.
    elif Tcold >= 18: 
        koppenClass = "A"
        # Tropical rainforest climate; average precipitation of at least 60 mm in every month.
        if Pdry >= 60:  
            koppenClass = koppenClass + "f"
        # Tropical monsoon climate; driest month (which nearly always occurs at or soon after the "winter" solstice for that side of the equator) with precipitation less than 60 mm (2.4 in), but at least
        else:
            if Pdry >= 100 - MAP / 25:  
                koppenClass = koppenClass + "m"
            #Tropical wet and dry or savanna climate; with the driest month having precipitation less than 60 mm (2.4 in) and less than
            else:
                koppenClass = koppenClass + "w"
                
    elif Thot > 10 and 0 < Tcold < 18:
        koppenClass = "C"
        if Psdry < 40 and Psdry < Pwwet / 3:
            koppenClass = koppenClass + "s"
        elif Pwdry < Pswet / 10:
            koppenClass = koppenClass + "w"
        else:
            koppenClass = koppenClass + "f"
        if Thot >= 22:
            koppenClass = koppenClass + "a"
        else:
            if Tmon10 >= 4:
                koppenClass = koppenClass + "b"
            elif 1 <= Tmon10 < 4:
                koppenClass = koppenClass + "c"
    elif Thot > 10 and Tcold <= 0:
        koppenClass = "D"
        if Psdry < 40 and Psdry < Pwwet / 3:
            koppenClass = koppenClass + "s"
        elif Pwdry < Pswet / 10:
            koppenClass = koppenClass + "w"
        else:
            koppenClass = koppenClass + "f"
        if Thot >= 22:
            koppenClass = koppenClass + "a"
        else:
            if Tmon10 >= 4:
                koppenClass = koppenClass + "b"
            elif Tcold < -38:
                koppenClass = koppenClass + "d"
            else:
                koppenClass = koppenClass + "c"
    elif Thot <= 10:
        koppenClass = "E"
        if Thot > 0:
            koppenClass = koppenClass + "T"
        else:
            koppenClass = koppenClass + "F"

    koppenDict = {
        "Af": 1,"Am": 2,"Aw": 3,
        "BWh": 4, "BWk": 5, "BSh": 6, "BSk": 7,
        "Csa": 8, "Csb": 9, "Csc": 10, "Cwa": 11, "Cwb": 12, "Cwc": 13, "Cfa": 14, "Cfb": 15, "Cfc": 16,
        "Dsa": 17, "Dsb": 18, "Dsc": 19, "Dsd": 20, "Dwa": 21, "Dwb": 22, "Dwc": 23, "Dwd": 24, "Dfa": 25, "Dfb": 26,
        "Dfc": 27, "Dfd": 28,
        "ET": 29, "EF": 30
    }
    koppenClass
    knu = koppenDict[koppenClass]
    return knu #koppenClass
```

```python
def custom_ufunc_along_x(a, b, c):
    if np.isnan(a).any():
        return 0
    else:
        # Example operation: multiply and sum along the 'x' dimension
        return koppen_beck(a,b,c) # Assuming 'x' is the last dimension
        #return a.sum()
```
