```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import ccf

# Sample time series data
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Calculate the Cross-Correlation Function
ccf_values = ccf(x, y)

# Plot the CCF
lags = np.arange(len(ccf_values))
plt.figure(figsize=(10, 5))
plt.bar(lags, ccf_values, width=0.1)
plt.xlabel('Lags')
plt.ylabel('Cross-Correlation')
plt.title('Cross-Correlation Function')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import grangercausalitytests

# Sample time series data
np.random.seed(0)
n = 100
x = np.random.randn(n)
y = np.random.randn(n) + 0.5 * np.roll(x, 1)  # Create some causality from x to y

# Combine the series into a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Perform Granger Causality test
max_lag = 4  # Define the maximum number of lags to test
test_result = grangercausalitytests(data[['y', 'x']], max_lag, verbose=True)

# Plot the data for visualization
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, label='x (Time Series 1)')
plt.title('Time Series x')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y, label='y (Time Series 2)')
plt.title('Time Series y')
plt.legend()

plt.tight_layout()
plt.show()
```
